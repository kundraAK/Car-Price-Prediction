```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Lasso
```

```python
carPrice_dataset = pd.read_csv('car data.csv')
```

```python
carPrice_dataset.shape
```

```python
carPrice_dataset.head()
```

```python
carPrice_dataset.isnull().sum()
```

```python
carPrice_dataset.describe()
```

```python
carPrice_dataset.Fuel_Type.value_counts()
```

```python
carPrice_dataset.Seller_Type.value_counts()
```

```python
carPrice_dataset.Transmission.value_counts()
```

```python
carPrice_dataset.replace({'Fuel_Type':{'Petrol':0, 'Diesel':1, 'CNG':2}},inplace=True)
```

```python
carPrice_dataset.replace({'Seller_Type': {'Dealer':0, 'Individual':1}},inplace=True)
```

```python
carPrice_dataset.replace({'Transmission': {'Manual':0, 'Automatic':1}},inplace=True)
```

```python
carPrice_dataset.head()
```

```python
X = carPrice_dataset.drop(['Car_Name', 'Selling_Price'],axis=1)
Y = carPrice_dataset['Selling_Price']
```

```python
Train_X, Test_X, Train_Y, Test_Y = train_test_split(X,Y,test_size=0.2, random_state=2)
```

```python
model1 = LinearRegression()
```

```python
model1.fit(Train_X, Train_Y)
```

```python
Training_pred = model1.predict(Train_X)
```

```python
print(metrics.r2_score(Training_pred, Train_Y))
```

```python
Testing_pred = model1.predict(Test_X)
```

```python
print(metrics.r2_score(Testing_pred, Test_Y))
```

```python
model2 = Lasso()
```

```python
model2.fit(Train_X, Train_Y)
```

```python
Training_pred = model2.predict(Train_X)
```

```python
print(metrics.r2_score(Training_pred, Train_Y))
```

```python
Testing_pred = model2.predict(Test_X)
```

```python
print(metrics.r2_score(Testing_pred, Test_Y))
```
